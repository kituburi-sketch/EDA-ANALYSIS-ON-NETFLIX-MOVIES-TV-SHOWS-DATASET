###  Scatter Plot: Country vs Movie Duration

```python
import matplotlib.pyplot as plt
import pandas as pd

# Keep only movies (since TV shows have 'Seasons' not 'minutes')
df_movies = df[df['type'] == 'Movie'].copy()

# Clean the 'duration' column (convert '90 min' → 90)
df_movies['duration_min'] = df_movies['duration'].str.replace(' min', '', regex=False)

# Drop rows where duration or country is missing
df_movies = df_movies.dropna(subset=['country', 'duration_min'])

# Convert duration to numeric safely
df_movies['duration_min'] = pd.to_numeric(df_movies['duration_min'], errors='coerce')

# Keep only top 10 countries for better visualization
top10_countries = df_movies['country'].value_counts().head(10).index
df_top10 = df_movies[df_movies['country'].isin(top10_countries)]

# Scatter plot
plt.figure(figsize=(10,6))
plt.scatter(df_top10['country'], df_top10['duration_min'], alpha=0.6, color='teal')
plt.xlabel('Country')
plt.ylabel('Duration (minutes)')
plt.title('Scatter Plot: Country vs Movie Duration')
plt.xticks(rotation=45)
plt.show()
```

###  LINE PLOT: Content Added Over the Years: Movies vs TV Shows

```python
# Ensure 'date_added' is datetime
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')
df['year_added'] = df['date_added'].dt.year

# Group by year and type
trend = df.groupby(['year_added', 'type']).size().reset_index(name='count')

plt.figure(figsize=(10,5))
sns.lineplot(data=trend, x='year_added', y='count', hue='type', marker='o')
plt.title('Content Added Over the Years: Movies vs TV Shows')
plt.xlabel('Year Added')
plt.ylabel('Number of Titles')
plt.legend(title='Type')
plt.show()
```

###  BOX PLOT:Distribution of Movie Durations on Netflix

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Filter only movies (since TV shows have "Seasons")
df_movies = df[df['type'] == 'Movie'].copy()

# Clean 'duration' column — remove "min" and convert to numeric
df_movies['duration_min'] = df_movies['duration'].str.replace(' min', '', regex=False)
df_movies['duration_min'] = pd.to_numeric(df_movies['duration_min'], errors='coerce')

# Drop missing values
df_movies = df_movies.dropna(subset=['duration_min'])

# Create boxplot
plt.figure(figsize=(6,5))
sns.boxplot(data=df_movies, y='duration_min', color='skyblue')
plt.title('Distribution of Movie Durations on Netflix')
plt.ylabel('Duration (minutes)')
plt.show()
```
